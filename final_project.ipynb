```python
#3. Low Level Data Understanding :
#a. Find count of unique values in location column.
# Read the data with proper delimiter and header
df = pd.read_csv('https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv'
, delimiter=",", header=0)

# Now you can perform the analysis tasks, such as finding the count of unique values in the 'location' column
unique_location_count = df['location'].nunique()
print("Count of unique values in the 'location' column:", unique_location_count)
```

```python
# b. Find which continent has maximum frequency using values 

# Find the continent with the maximum frequency using value counts
continent_max_frequency = df['continent'].value_counts().idxmax()

# Find the maximum frequency count
max_frequency_count = df['continent'].value_counts().max()

# Print the continent with maximum frequency and the maximum frequency count
print(f"The continent with the maximum frequency is '{continent_max_frequency}' with a count of {max_frequency_count}.")
```

```python
# c. Find maximum & mean value in 'total_cases'.
df['total_cases'].max()
df['total_cases'].mean()
```

```python
#d. Find 25%,50% & 75% quartile value in 'total_deaths'.
df['total_deaths'].quantile([0.25, 0.5, 0.75])
```

```python
# e. Find which continent has maximum'human_development_index'.
# Group the data by 'continent' and find the maximum 'human_development_index' within each continent
continent_max_hdi = df.groupby('continent')['human_development_index'].max()

# Find the continent with the maximum 'human_development_index'
continent_with_max_hdi = continent_max_hdi.idxmax()

# Find the maximum 'human_development_index' value
max_hdi_value = continent_max_hdi.max()

# Print the continent with maximum 'human_development_index' and the maximum value
print(f"The continent with the maximum 'human_development_index' is '{continent_with_max_hdi}' with a value of {max_hdi_value}.")
```

```python
#f. Find which continent has minimum 'gdp_per_capita'.
# Group the data by 'continent' and find the minimum 'gdp_per_capita' within each continent
continent_min_gdp = df.groupby('continent')['gdp_per_capita'].min()

# Find the continent with the minimum 'gdp_per_capita'
continent_with_min_gdp = continent_min_gdp.idxmin()

# Find the minimum 'gdp_per_capita' value
min_gdp_value = continent_min_gdp.min()

# Print the continent with minimum 'gdp_per_capita' and the minimum value
print(f"The continent with the minimum 'gdp_per_capita' is '{continent_with_min_gdp}' with a value of {min_gdp_value}.")
```

```python
# #4. Filter the dataframe with only this columns
# ['continent','location','date','total_cases','total_deaths','gdp_per_ca
# pita','
# human_development_index'] and update the data frame.
# List of columns to keep in the DataFrame
columns_to_keep = ['continent', 'location', 'date', 'total_cases', 'total_deaths', 'gdp_per_capita', 'human_development_index']

# Filter the DataFrame to include only the specified columns
filtered_df = df[columns_to_keep]

# Update the original DataFrame with the filtered DataFrame
df = filtered_df
df.head()
```

```python
# Data Cleaning
#  a. Remove all duplicates observations
df.drop_duplicates()
```

```python
#b. Find missing values in all columns
df.isna().sum()
```

```python
#c. Remove all observations where continent column value is missing
df.dropna(subset=['continent'])
```

```python
# d. Fill all missing values with 0
df.fillna(0)
```

```python
#6. Date time format :
#  a. Convert date column in datetime format using
# pandas.to_datetime
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y')  # Adjust the date format as needed
```

```python
#b. Create new column month after extracting month data from date column
df['month'] = df['date'].dt.month
df.head()
```

```python
# 7. Data Aggregation:
#  a. Find max value in all columns using groupby function on
# 'continent'
#  column
#  Tip: use reset_index() after applying groupby
df.groupby('continent').max().reset_index()
df.head()
```

```python
# b. Store the result in a new dataframe named 'df_groupby'.
#  (Use df_groupby dataframe for all further analysis)
df.groupby('continent').max().reset_index()
```

```python
# #8. Feature Engineering :
#  a. Create a new feature 'total_deaths_to_total_cases' by ratio of
#  'total_deaths' column to 'total_cases'
# Group the DataFrame by 'continent' and find the maximum value in each group
df_groupby = df.groupby('continent').max().reset_index()

# Create the new feature 'total_deaths_to_total_cases'
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
```

```python
# 9. Data Visualization :
#  a. Perform Univariate analysis on 'gdp_per_capita' column by
# plotting
#  histogram using seaborn dist plot.
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style for the plots (optional)
sns.set(style="whitegrid")

# Create a histogram using Seaborn's distplot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
sns.distplot(df_groupby['gdp_per_capita'], kde=False, bins=20, color='blue')  # Adjust the number of bins and color

# Add labels and a title
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')
plt.title('Histogram of GDP per Capita')

# Show the histogram
plt.show()
```

```python
# b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'

# Set the style for the plots (optional)
sns.set(style="whitegrid")

# Create a scatter plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
sns.scatterplot(x='gdp_per_capita', y='total_cases', data=df_groupby, hue='continent', palette='viridis')  # Adjust the hue and palette if needed

# Add labels and a title
plt.xlabel('GDP per Capita')
plt.ylabel('Total Cases')
plt.title('Scatter Plot of Total Cases vs GDP per Capita')

# Show the scatter plot
plt.show()
```

```python
#c. Plot Pairplot on df_groupby dataset.
# Set the style for the plots (optional)
sns.set(style="ticks")

# Create a pairplot
sns.pairplot(df_groupby, hue='continent', palette='viridis')  # Adjust hue and palette as needed

# Show the pairplot
plt.show()
```

```python
# # d. Plot a bar plot of 'continent' column with 'total_cases' .
#  Tip : using kind='bar' in seaborn catplot
sns.set(style="whitegrid")

# Create a bar plot using Seaborn's catplot
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
sns.catplot(x='continent', y='total_cases', data=df_groupby, kind='bar', palette='muted', aspect=2)

# Add labels and a title
plt.xlabel('Continent')
plt.ylabel('Total Cases')
plt.title('Bar Plot of Total Cases by Continent')

# Show the bar plot
plt.show()
```
